```python
import bs4
import urllib.request as req
import numpy as np
import re
import pandas as pd
import requests
```

```python
ndtv='https://gadgets.ndtv.com/mobiles/all-brands'
```

```python
resp=requests.get(ndtv).content
```

```python
resp
```

```python

ndtv_html=bs4.BeautifulSoup(resp,'html.parser')
```

```python
company_html=ndtv_html.find_all('ul',class_='clearfix margin_t20')
```

```python
company_html=company_html[1].select('li')
```

```python
company_url={}
company_html[0].select_one('a')
```

```python
for i in company_html:
    print(i.select('a'))
```

```python
company_details={'company':[],'company_url':[],'company_label':[]}
for company_detail in company_html:
    
    company_details['company'].append(company_detail.select_one('img')['title'])
    company_details['company_url'].append(company_detail.select_one('a')['href'])
    company_details['company_label'].append(company_detail.select_one('img')['data-original'])

company_details=pd.DataFrame(company_details)
```

```python
company_details.to_csv('company_details.csv')
```

```python
count=1
phones=mobiles_urls['mobile'].values
final=[]
for phone in phones:
    temp=bs4.BeautifulSoup(requests.get(phone).content)
    specs={}
    details=temp.select('div#specs div._gry-bg._spctbl._ovfhide table tbody tr')
    for detail in details:
        try:
            if detail.select('td')[0].text in specs:
                specs[detail.select('td')[0].text]=specs[detail.select('td')[0].text]+ ' || ' +detail.select('td')[1].text

            else:
                specs[detail.select('td')[0].text]=detail.select('td')[1].text
            #print(detail.select('td')[0].text,': ', detail.select('td')[1].text)
        except Exception as e:
            continue
    final.append(specs)
    print(count)
    count=count+1
    
```

```python
pd.DataFrame( final).to_csv('specs.csv')
```
