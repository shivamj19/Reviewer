```python
import bs4
import urllib.request as req
import numpy as np
import re
import pandas as pd
import requests
from collections import defaultdict
```

```python
gsm_arena='https://www.gsmarena.com/'
```

```python
gsm_response=req.urlopen(gsm_arena)
```

```python
gsm_html=bs4.BeautifulSoup(gsm_response)
```

```python
companies=gsm_html.select('div.brandmenu-v2.light.l-box.clearfix li')
```

```python
companies_name_list=[]
for company in companies:
    companies_name_list.append(company.text.lower())
```

```python
company_url={}
for company in companies:
    if company not in company_url:
        company_url[company.find('a').text.lower()]='https://www.gsmarena.com/'+company.find('a')['href']
```

```python
#for company in companies_name_list:
 #   if company in company_url.keys():
  #      print(company,company_url[company])
```

```python
all_company_page_url=defaultdict(list)
for company in companies_name_list:
    company_page_response=req.urlopen(company_url[company])
    company_page=bs4.BeautifulSoup(company_page_response)
    if company not in all_company_page_url:
        all_company_page_url[company].append(company_url[company])
        for prod in company_page.select('div.review-nav.pullNeg.col.pushT10 div.nav-pages a'):
            all_company_page_url[company].append('https://www.gsmarena.com/'+prod['href'])
```
